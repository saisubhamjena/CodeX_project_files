```python
csv=pd.read_csv("fact_survey_responses.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
l1=np.random.normal(30,7,1009)
l2=np.random.normal(75,10,4288)
l3=np.random.normal(125,10,3142)
l4=np.random.normal(175,10,1561)



plt.figure(figsize=(20,5))

plt.subplot(1,4,1)
sns.histplot(x=l1)

plt.subplot(1,4,2)
sns.histplot(x=l2)

plt.subplot(1,4,3)
sns.histplot(x=l3)

plt.subplot(1,4,4)
sns.histplot(x=l4)
```

```python
total=list(l1)+list(l2)+list(l3)+list(l4)
```

```python
df=pd.DataFrame(data=total)
df.to_csv("simulation_gussian.csv")
```

```python
plt.figure(figsize=(20,5))
sns.set_style('white')
sns.histplot(x=total,kde=True)
plt.show()
```
